```python
#First, we need to query CTP data from Parse. Since we will also need data for M1 - M6 cancellation. Let's get them 
#altogether
# url = https://parse-production.learnwithhomer.com/
import json, httplib, urllib
import datetime, time
def get_nonpathway_user(seconds_since_epoch, seconds_end_epoch, page_size = 10000):
    url = 'localhost:1337'
    connection = httplib.HTTPConnection(url)
    page = 0
    first_loop = True
    page_data = []
    while first_loop or len(page_data) > 0:  
        first_loop = False
        params = urllib.urlencode({
            "skip": page * page_size,
            "limit": page_size,
            'where':
            json.dumps({
                "iapPurchaseDate":{"$gte": seconds_since_epoch,"$lt":seconds_end_epoch}
                #"reportingStatus":"PAID_1"                
            })
        })
        connection.connect()
        connection.request('GET', '/parse/classes/IAPSubscription?%s' % params, '', {
            'X-Parse-Application-Id': '8nyG46QswpUa2YidGVrZj5S9WupfmPwmNbCFEvOL',
            'X-Parse-REST-API-Key': 'OTCQ9MT93dtdkIOcNT2mMG0bZ668RZ9ttdjAEVfC',
            'X-Parse-Master-Key':'tYKiwqU7js4KrUcmNm4lCUEhmiRQc6jTffiYo1JV'
         })
        result = connection.getresponse()
        print result.status
        try:
            page_data = json.loads(result.read())['results']
        except Exception, e:
            page_data = []
        page += 1
        yield page_data 
```
